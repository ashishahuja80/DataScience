# Poisson Distribution

Poisson Distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event. The Poisson distribution can also be used for the number of events in other specified intervals such as distance, area or volume. [SOURCE](https://en.wikipedia.org/wiki/Poisson_distribution)

![Poisson](./poisson_def.png)

### Properties of Poisson Distribution

1. Poisson Distribution is a Discrete Distribution.  
1. Events are independent of each other.
1. The average number of events per interval (also called rate) is constant.
1. The occurrence of one event does not affect the probability another event.
1. Events are not simultaneous.
1. If the number of events per unit time follows a Poisson distribution, then the amount of time between events follows the exponential distribution, which is a continuous distribution.



```python
from numpy import random
from scipy.stats import poisson
import numpy as np
#lam - rate or known number of occurences e.g. 2 

#size - The shape of the returned array.

x = random.poisson(lam=2, size=20)

print(x)
```

### Visualization of Poisson Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

f, ax = plt.subplots(figsize=(12,8))
sns.distplot(random.poisson(lam=2, size=1000), kde=False)

#sns.lineplot()

plt.show()
```

```python
mu_list=[0,1,2,3,4,5,6]
#mu=2
#def poisson_data(mu_list):
fig, ax = plt.subplots(figsize=(12,8))

for mu in mu_list:
    x = np.arange(poisson.ppf(0.01, mu),
                  poisson.ppf(0.99, mu))
    ax.plot(x, poisson.pmf(x, mu), 'bo')
    sns.lineplot(x,poisson.pmf(x, mu),marker=True,label='lambda = ' + str(mu))
    
plt.xlabel('Number of Events')
plt.ylabel('Probability P(X=k)')
```
